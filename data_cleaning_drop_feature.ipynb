```python
import pandas as pd # load dataset 
import seaborn as sns # generate heat map
import matplotlib.pyplot as plt
tableA = pd.read_csv('4. Dataset Aggregation File/MeterA_aggregation.csv', header='infer') # read MeterA csv file
tableA.drop(tableA.filter(regex="Unname"),axis=1, inplace=True)
tableB = pd.read_csv('4. Dataset Aggregation File/MeterB_aggregation.csv', header='infer') # read MeterB csv file
tableB.drop(tableB.filter(regex="Unname"),axis=1, inplace=True)
tableC = pd.read_csv('4. Dataset Aggregation File/MeterC_aggregation.csv', header='infer') # read MeterC csv file
tableC.drop(tableC.filter(regex="Unname"),axis=1, inplace=True)
```

```python
corrmatA = tableA.corr() # get correlation
plt.figure(figsize=(10,6)) 
sns.heatmap(corrmatA, annot=True)
```

```python
#DROP flow velocity only

tableA.drop(f'Average Flow Velocity', axis=1, inplace=True)

corrmatA = tableA.corr() # get correlation
plt.figure(figsize=(10,4)) 
sns.heatmap(corrmatA, annot=True)
#tableB.drop(f'Flow Velocity{i+1}', axis=1, inplace=True)
```

```python
corrmatB = tableB.corr() # get correlation
plt.figure(figsize=(10,6)) 
sns.heatmap(corrmatB, annot=True)
```

```python
#drop avg turbulence, signal quality

tableB.drop(f'Average Turbulence', axis=1, inplace=True)
#tableB.drop(f'Average speed of sound', axis=1, inplace=True)
#tableB.drop(f'Average Signal Strength', axis=1, inplace=True)
tableB.drop(f'Average Signal Quality', axis=1, inplace=True)

corrmatB = tableB.corr() # get correlation
plt.figure(figsize=(10,6)) 
sns.heatmap(corrmatB, annot=True)
```

```python
corrmatC = tableC.corr() # get correlation
plt.figure(figsize=(10,6)) 
sns.heatmap(corrmatC, annot=True)
```

```python
#drop transit time ,avg speed of sound

#tableC.drop(f'Symmetry', axis=1, inplace=True)
#tableC.drop(f'Crossflow', axis=1, inplace=True)
tableC.drop(f'Average Speed of Sound', axis=1, inplace=True)
tableC.drop(f'Average Transit Time', axis=1, inplace=True)

corrmatC = tableC.corr() # get correlation
plt.figure(figsize=(7,4)) 
sns.heatmap(corrmatC, annot=True)

```

```python
# write data to csv file 

corrmatC= corrmatC.reset_index(drop=True)
corrmatC.to_csv('5. Dataset Feature Dropping/MeterC_featureSelection.csv', index = True)

corrmatB= corrmatB.reset_index(drop=True)
corrmatB.to_csv('5. Dataset Feature Dropping/MeterB_featureSelection.csv', index = True)

corrmatA= corrmatA.reset_index(drop=True)
corrmatA.to_csv('5. Dataset Feature Dropping/MeterA_featureSelection.csv', index = True)

print("done")
```
