**Get Z score for dataset A**

```python
%matplotlib inline 
tableA2 = tableA.drop(['Class Attruibute'],axis=1) # drop Class Attribute column 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
Z = (tableA2-tableA2.mean())/tableA2.std() # get z score
Z
```

**Identify outliers based on Z score for dataset A**

```python
print('Number of rows before discarding outliers = %d' % (Z.shape[0])) # drop outlier rows 
outlier_index = []
for index, row in Z.iterrows(): # append the index if z score is higher than 3 or lower than -3 
    for i in row:
        if i > 3 or i < -3:
            outlier_index.append(index)
            
set_outlier = set(outlier_index)
outlier_index_list = list(set_outlier)
print(outlier_index_list)
count = 0

list_outlier = [tableA.index[i] for i in outlier_index_list]
tableA_no_outlier = tableA.drop(list_outlier) # drop the outlier row 
print(tableA_no_outlier)

Z2 = Z.loc[((Z > -3).sum(axis=1)==36) & ((Z <= 3).sum(axis=1)==36),:]
print('Number of rows after discarding outliers = %d' % (Z2.shape[0]))
```

**Transform new dataset A to csv**

```python
tableA_no_outlier = tableA_no_outlier.reset_index(drop=True)
tableA_no_outlier.to_csv('MeterA_no_outlier.csv', index = True)
```
