```python
import pandas as pd 
tableA = pd.read_csv('Dataset CSV File/MeterA_mod.csv', header='infer') # read MeterA csv file
tableB = pd.read_csv('Dataset CSV File/MeterB_mod.csv', header='infer') # read MeterB csv file
tableC = pd.read_csv('Dataset CSV File/MeterC_mod.csv', header='infer') # read MeterC csv file
```

```python
import numpy as np

tableA = tableA.replace('?',np.NaN) # Check if tableC contains null value

print('Number of instances = %d' % (tableA.shape[0]))
print('Number of attributes = %d' % (tableA.shape[1]))

print('Number of missing values:')
for col in tableA.columns:
    print('\t%s: %d' % (col,tableA[col].isna().sum()))
```

```python
tableB= tableB.replace('?',np.NaN) # Check if tableB contains null value

print('Number of instances = %d' % (tableB.shape[0]))
print('Number of attributes = %d' % (tableB.shape[1]))

print('Number of missing values:')
for col in tableB.columns:
    print('\t%s: %d' % (col,tableB[col].isna().sum()))
```

```python
tableC = tableC.replace('?',np.NaN) # Check if tableC contains null value

print('Number of instances = %d' % (tableC.shape[0]))
print('Number of attributes = %d' % (tableC.shape[1]))

print('Number of missing values:')
for col in tableC.columns:
    print('\t%s: %d' % (col,tableC[col].isna().sum()))
```

```python
%matplotlib inline
# tableA.boxplot(figsize=(10,3))
tableA2 = tableA.drop(['Class Attruibute'],axis=1)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
Z = (tableA2-tableA2.mean())/tableA2.std() # get z score
Z
```

```python
print('Number of rows before discarding outliers = %d' % (Z.shape[0])) # drop outlier rows 
outlier_index = []
for index, row in Z.iterrows():
    for i in row:
        if i > 3 or i < -3:
            outlier_index.append(index)
            
set_outlier = set(outlier_index)
outlier_index_list = list(set_outlier)
print(outlier_index_list)
count = 0

list_outlier = [tableA.index[i] for i in outlier_index_list]
tableA_no_outlier = tableA.drop(list_outlier)
print(tableA_no_outlier)

Z2 = Z.loc[((Z > -3).sum(axis=1)==36) & ((Z <= 3).sum(axis=1)==36),:]
print('Number of rows after discarding outliers = %d' % (Z2.shape[0]))
```

```python
tableA_no_outlier = tableA_no_outlier.reset_index(drop=True)
tableA_no_outlier.to_csv('MeterA_no_outlier.csv', index = True)
```
